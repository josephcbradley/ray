```python
from pathlib import Path
import subprocess
```

```python
reqs_path = Path("reqs")

reqs_to_process: list[Path] = []

for path in reqs_path.rglob("*.in"):
    if path.stem == "core":
        continue
    else:
        reqs_to_process.append(path)
```

```python

core_req = reqs_path / Path("core.in")
if not core_req.is_file():
    raise FileNotFoundError("Could not find the core file.")

outputs_dir = Path("outputs")
outputs_dir.mkdir(exist_ok=True)
pyver = "3.12"
for req in reqs_to_process:

    result = subprocess.run(
        f"uv pip compile {core_req} {req} --python-version {pyver} -o {outputs_dir / Path(f"{req.stem}.out")}".split(" ")
    )
```

```python
def nohash(text: str) -> bool:
    if "#" in text:
        return False 
    else:
        return True

combined_str = []

for reqpath in Path("outputs").iterdir():
    with open(reqpath, "r") as infile: 
        lines = infile.readlines()
        combined_str.extend(lines)
print(f"Initial length: {len(combined_str)}")
combined_str = list(set([line for line in combined_str if nohash(line)]))
print(f"Final length: {len(combined_str)}")
print(combined_str)
```

```python
with open("all_pkgs.txt", "w") as outfile:
    outfile.write("".join(combined_str))
```
